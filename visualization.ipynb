```python
import pandas as pd
import csv
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('application_train.csv')

sns.set(style="ticks", color_codes=True)
#plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
```
